```python
#Imports
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import SpectralClustering
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

## Finding Optimal Number of Clusters
### Elbow Criterion and Silhouette Coefficient Methods:

```python
# Load data set
dataset = pd.read_csv("loans_train_prepared-1.csv")
data = np.array(dataset.values, dtype=float)
```

```python
def kmean_optimal_k_search(no_of_clusters):
    sum_squared_error = {} 
    coefficient = []

    for no_clusters in range(2, no_of_clusters):
        kmeans = KMeans(n_clusters=no_clusters, n_init=10, max_iter=300, random_state=6).fit(data)
        sum_squared_error[no_clusters] = kmeans.inertia_
        label = kmeans.labels_
        sil_coef = silhouette_score(data, label, metric='euclidean')
        coefficient.append(sil_coef)
        print("For number of clusters = {} the coefficint is {}".format(no_clusters, sil_coef))
    fig = plt.figure(figsize=(16,8))
    ax1 = fig.add_subplot(1,2,1)
    plt.plot(list(sum_squared_error.keys()), list(sum_squared_error.values()))
    plt.xlabel("Number of Clusters")
    plt.ylabel("Sum Squared Error")
    plt.grid()

    ax1 = fig.add_subplot(1,2,2)
    plt.plot(coefficient)
    plt.xlabel("Number of Clusters")
    plt.ylabel("Silhouette Coefficient")
    plt.grid()
    plt.show()
```

```python
kmean_optimal_k_search(614)
```

The purpose of the above experiment was to ascertain that when the number of clusters become equal to the size of the sample the sum of squared error does reach 0, and the Silhouette coefficient also gets closer to 0. The above plots show the same. Moreover, we can also observe that experimenting with a value of 'k' equal to the number of samples each plot suggests a very large value for the optimal number of 'k'. In this case, the optimal value of 'k' is 207 clusters. However, we only have 614 observations in our dataset, hence the optimal value of 'k' equal to 207 doesn't seem to be a good choice. Therefore, we must depend on domain knowledge to decide on an optimal number of clusters. As the data in hand deals with loan, therefore it is likely that it will not have many clusters, hence I try to find optimal number of clusters by trying values of k in range (2, 10) inclusive.

```python
kmean_optimal_k_search(11)
```

#### If we jointly interpret the above two plots, we can conclude that optimal number of clusters (k) is equal to 8.

```python
import warnings
warnings.filterwarnings('ignore')
#Let's first reduce the dimension using t-SNE
# Let's first reload the data 
dataset = pd.read_csv("loans_train_prepared-1.csv")
data = np.array(dataset.values, dtype=float)
print("Shape of orginal dataset {}".format(data.shape))

# We have 16 features/variable in our dataset, let's use t-SNE to reduce the dimension to 2
# We don't need to scale the data as it is already in a good shape
tsne = TSNE(n_components=2)
data = tsne.fit_transform(data)
print("Shape of transformed dataset {}".format(data.shape))
```

```python
def spectral_clustering(no_of_clusters):
    dataset = pd.DataFrame(data, columns=["feature_1", "feature_2"])
    spectral_clustering = SpectralClustering(n_clusters=no_of_clusters, eigen_solver= "arpack", n_init=10, random_state=6).fit(dataset)
    dataset["spectral_clustering_2"] = spectral_clustering.labels_

    fig = plt.figure(figsize=(16,8))
    ax1 = fig.add_subplot(1,2,1)
    plt.scatter(dataset["feature_1"], dataset["feature_2"], c=dataset["spectral_clustering_2"])
    plt.grid()
    plt.title("Clusters Using Spectral Clustering with arpack Eigen Solver")

    dataset = pd.DataFrame(data, columns=["feature_1", "feature_2"])
    spectral_clustering = SpectralClustering(n_clusters=no_of_clusters, eigen_solver= "lobpcg", n_init=10, random_state=6).fit(dataset)
    dataset["spectral_clustering_2"] = spectral_clustering.labels_

    ax1 = fig.add_subplot(1,2,2)
    plt.scatter(dataset["feature_1"], dataset["feature_2"], c=dataset["spectral_clustering_2"])
    plt.grid()
    plt.title("Clusters Using Spectral Clustering with lobpcg Eigen Solver")
```

```python
spectral_clustering(2)
```

The above plots show that spectral clustering with two clusters doesn't show good results as clusters are not contiguous. The above plots demonstrate that there could be four clusters. Therefore, let's try spectral clustering with n_cluster = 4.

```python
spectral_clustering(4)
```

The above plots demonstrate that even if we increase the number of clusters from 2 to 4, the spectral clustering technique using different eigen solver is still not able to come up with good cluster organization. 

```python
spectral_clustering(8)
```
