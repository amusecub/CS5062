### Distribution of Numeric Attributes

```python
subset_data = data[["age", "balance", "duration", "pdays"]]
subset_data.hist()
```

```python
subset_data.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
```

### Conclusions
Histograms and box graphs for the following numeric attributes are plotted: age, balance, duration, and days since last contact (pdays). Histogram corresponding to the age attribute shows that the distribution of age is kind of skewed normal distribution. The histogram corresponding to the balance attribute shows that the attribute balance exhibits exponential distribution. Similarly, contact duration also exhibits exponential distribution. The pdays attribute exhibits random distribution.
The box plots also provide reassurance about the comments made about the distribution of the attributes considering the histograms. Moreover, it can also be observed from the box plots that the age attribute has minimum number of outliers, and on the other hands the other shown attributes demonstrate relatively large number of outliers.
